```python
import matplotlib.pyplot as plt
```

```python
epochs = [1, 2, 3, 4, 5]
```

```python
train_loss = [
    0.29091294572625787,
    0.07765235238782833,
    0.03488803409365651,
    0.007497639238885804,
    0.0024330466421481015,
]
```

```python
test_loss = [
    0.08791896360344253,
    0.05930741100019077,
    0.0897792825999204,
    0.07531580289105477,
    0.07241666069785424,
]
```

```python
fig, ax = plt.subplots(1, figsize=(8, 6))

ax.plot(epochs, train_loss, marker="o", label="Train Loss", color="blue")
ax.plot(epochs, test_loss, marker="o", label="Test Loss", color="orange")

# Annotation
for i in range(len(epochs)):
    ax.text(
        epochs[i],
        train_loss[i] + 0.01,
        f"{train_loss[i]:.4f}",
        ha="center",
        va="bottom",
        fontsize=9,
        color="blue",
    )
    ax.text(
        x=epochs[i],
        y=test_loss[i] + 0.005,
        s=f"{test_loss[i]:.4f}",
        ha="center",
        va="bottom",
        fontsize=9,
        color="orange",
    )

ax.set_xlabel("Época")
ax.set_xticks(epochs)

ax.set_ylabel("Loss")

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_title(
    "Loss de treinamento e teste ao longo das épocas",
    fontsize=14,
    y=1.1,
)

plt.legend()

plt.savefig(
    "train_and_test_loss.png",
    transparent=False,
    dpi=300,
    # bbox_inches="tight",
)

plt.show()
```
